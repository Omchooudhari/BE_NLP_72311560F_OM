```python
import nltk
from nltk.tokenize import word_tokenize, TweetTokenizer, MWETokenizer
from nltk.tokenize.treebank import TreebankWordTokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('wordnet')
```

```python
# input
text = "Hello! I am learning NLP. This is a great experience :) #AI"
print(text)
```

```python
# Tokenization
print("\n--- Whitespace Tokenization ---")
print(text.split())

print("\n--- Punctuation Tokenization ---")
print(word_tokenize(text))
```

```python
tb = TreebankWordTokenizer()
tw = TweetTokenizer()                   # tweet tokenizer (handles hashtags, emojis,....)

print("\n--- Treebank ---")
print(tb.tokenize(text))

print("\n--- Tweet ---")
print(tw.tokenize(text))
```

```python
# mwe
mwe = MWETokenizer([('New', 'York'), ('machine', 'learning')])
text2 = "I live in New York and love machine learning"

print("\n--- MWE ---")
print(mwe.tokenize(text2.split()))
```

```python
# Stemming
words = word_tokenize(text)

ps = PorterStemmer()
ss = SnowballStemmer("english")

print("\n--- Porter ---")
print([ps.stem(w) for w in words])

print("\n--- Snowball ---")
print([ss.stem(w) for w in words])
```

```python
# lemmatization 
lem = WordNetLemmatizer()

print("\n--- Lemmatization ---")
print([lem.lemmatize(w) for w in words])
```
